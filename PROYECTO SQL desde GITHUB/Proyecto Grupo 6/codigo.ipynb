```python
#Vamos a crear la df relacionada a Vertical
vertical = pd.DataFrame()
vertical["Vertical_id"] = df["Vertical_id"]
vertical["Campus"] = Campus
vertical["Promocion"] = Promocion
vertical = vertical.head(4)
```

```python
#Para esta dataframe, cambiamos valores a mano para adecuarla.
vertical.at[1, 'Vertical_id'] = 'DS_FEB'
vertical.at[3, 'Vertical_id'] = 'FS_FEB'
vertical.at[1, 'Promocion'] = 'Febrero'
vertical.at[3, 'Promocion'] = 'Febrero'
```

```python
#Creamos la columna de fechas y la metemos.
#vamos a cojer de las DF de alumnos las fechas de febrero y septiembre
Fecha_febrero = df_fs2.at[1,"Fecha_comienzo"]
Fecha_septiembre = df_ds1.at[1,"Fecha_comienzo"]
#aqui ya metemos las fechas en su lugar correspondiente.
vertical["Fecha_comienzo"] = vertical["Promocion"]
vertical.loc[vertical['Promocion'] == 'Septiembre', 'Fecha_comienzo'] = Fecha_septiembre
vertical.loc[vertical['Promocion'] == 'Febrero', 'Fecha_comienzo'] = Fecha_febrero
vertical
```

```python
#El formato de fecha de SQL es YYYY-MM-DD
#vamos a cambiarle, el tipo de columna a tipo datetime
vertical['Fecha_comienzo'] = pd.to_datetime(vertical['Fecha_comienzo'], format='%d/%m/%Y')

#ahora, cambiamos el formato de la fecha

vertical['Fecha_comienzo'] = vertical['Fecha_comienzo'].dt.strftime('%Y-%m-%d')
vertical
```

```python
#Tabla terminada, la guardamos.
ruta = 'data_limpiado/Vertical.csv'
vertical.to_csv(ruta, index=False)
```

```python
#Creamos la tabla estudiantes.

estudiantes_DS = pd.DataFrame()
estudiantes_FS = pd.DataFrame()
#Solo nos interesan columnas en particular, juntamos todas las filas de las 4 tablas por las columnas que queremos.


estudiantes_DS["Email"] = pd.concat([df_ds1["Email"], df_ds2["Email"]], axis=0, ignore_index=True)
estudiantes_DS["Vertical_id"] = pd.concat([df_ds1["Campus"], df_ds2["Campus"]], axis=0, ignore_index=True)

estudiantes_FS["Email"] = pd.concat([df_fs1["Email"], df_fs2["Email"]], axis=0, ignore_index=True)
estudiantes_FS["Vertical_id"] = pd.concat([df_fs1["Campus"], df_fs2["Campus"]], axis=0, ignore_index=True)

#Vamos a cambiar la vertical_id segun la fecha de comienzo, metemos la fecha de comienzo de cada estudiante.
estudiantes_DS["Fecha_comienzo"] = pd.concat([df_ds1["Fecha_comienzo"], df_ds2["Fecha_comienzo"]], axis=0, ignore_index=True)
estudiantes_FS["Fecha_comienzo"] = pd.concat([df_fs1["Fecha_comienzo"], df_fs2["Fecha_comienzo"]], axis=0, ignore_index=True)
```

```python
#Aqui, cambiamos, segun sean de Madrid o valencia, y segun su fecha de comienzo
estudiantes_DS.loc[(estudiantes_DS["Fecha_comienzo"] == Fecha_febrero), 'Vertical_id'] = 'DS_FEB'
estudiantes_DS.loc[(estudiantes_DS["Fecha_comienzo"] == Fecha_septiembre), 'Vertical_id'] = 'DS_SEP'
estudiantes_FS.loc[(estudiantes_FS["Fecha_comienzo"] == Fecha_febrero), 'Vertical_id'] = 'FS_FEB'
estudiantes_FS.loc[(estudiantes_FS["Fecha_comienzo"] == Fecha_septiembre), 'Vertical_id'] = 'FS_SEP'

```

```python
#Ultimos ajustes para adaptarla a nuestro esquema.
estudiantes_final = pd.concat([estudiantes_DS, estudiantes_FS], axis=0, ignore_index=True)
estudiantes_final = estudiantes_final.drop("Fecha_comienzo", axis=1)
estudiantes_final["Nombre"] = pd.concat([df_ds1["Nombre"], df_ds2["Nombre"],df_fs1["Nombre"], df_fs2["Nombre"]], axis=0, ignore_index=True)
estudiantes_final
```

```python
#ya esta terminada, la importamos.
ruta = 'data_limpiado/Estudiantes.csv'
estudiantes_final.to_csv(ruta, index=False)
```

```python
#Creamos la tablas proyectos segun vertical, para registrar las calificaciones de cada alumno.
proyectos_DS = pd.DataFrame()


# Añadimos las columnas de los emails correspondientes que van a ser las primary keys en cada tabla correspondiente por vertical
proyectos_DS['Email'] = estudiantes_DS["Email"]

# Añadimos todas las columnas de los diferentes proyectos y sus califaciones correspondientes
proyectos_DS["Proyecto_HLF"] = pd.concat([df_ds1["Proyecto_HLF"], df_ds2["Proyecto_HLF"]], axis=0, ignore_index=True)
proyectos_DS["Proyecto_EDA"] = pd.concat([df_ds1["Proyecto_EDA"], df_ds2["Proyecto_EDA"]], axis=0, ignore_index=True)
proyectos_DS["Proyecto_BBDD"] = pd.concat([df_ds1["Proyecto_BBDD"], df_ds2["Proyecto_BBDD"]], axis=0, ignore_index=True)
proyectos_DS["Proyecto_ML"] = pd.concat([df_ds1["Proyecto_ML"], df_ds2["Proyecto_ML"]], axis=0, ignore_index=True)
proyectos_DS["Proyecto_Deployment"] = pd.concat([df_ds1["Proyecto_Deployment"], df_ds2["Proyecto_Deployment"]], axis=0, ignore_index=True)

proyectos_DS

```

```python
#ya esta terminada, la importamos.
ruta = 'data_limpiado/proyectos_DS.csv'
proyectos_DS.to_csv(ruta, index=False)
```

```python
#Creamos la tablas proyectos segun vertical, para registrar las calificaciones de cada alumno.

proyectos_FS = pd.DataFrame()


# Añadimos las columnas de los emails correspondientes que van a ser las primary keys en cada tabla correspondiente por vertical

proyectos_FS['Email'] = estudiantes_FS["Email"]

# Añadimos todas las columnas de los diferentes proyectos y sus califaciones correspondientes
proyectos_FS["Proyecto_WebDev"] = pd.concat([df_fs1["Proyecto_WebDev"], df_fs2["Proyecto_WebDev"]], axis=0, ignore_index=True)
proyectos_FS["Proyecto_FrontEnd"] = pd.concat([df_fs1["Proyecto_FrontEnd"], df_fs2["Proyecto_FrontEnd"]], axis=0, ignore_index=True)
proyectos_FS["Proyecto_BackEnd"] = pd.concat([df_fs1["Proyecto_Backend"], df_fs2["Proyecto_Backend"]], axis=0, ignore_index=True)
proyectos_FS["Proyecto_React"] = pd.concat([df_fs1["Proyecto_React"], df_fs2["Proyecto_React"]], axis=0, ignore_index=True)
proyectos_FS["Proyecto_FullStack"] = pd.concat([df_fs1["Proyecto_FullStack"], df_fs2["Proyecto_FullStack"]], axis=0, ignore_index=True)

proyectos_FS
```
