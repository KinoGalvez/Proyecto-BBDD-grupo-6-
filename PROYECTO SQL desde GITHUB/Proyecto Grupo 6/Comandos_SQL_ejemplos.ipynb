## Una query mas avanzada para testear, tenemos varios TA, por alumno, pues para que salga en una sola columna los diferentes TA que tiene el correspondiente estudiante

SELECT  e.nombre,  string_agg(cd.nombre, ', ') AS TA


FROM estudiantes e


LEFT JOIN cuerpo_docente cd ON e.vertical_id = cd.vertical_id

WHERE cd.rol = 'TA'


GROUP BY e.nombre;
